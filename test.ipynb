```python
from pymongo import MongoClient
from pprint import pprint

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")
db = client["ev_database"]
```

```python
document = db["vehicle_data"].find_one()
pprint(document)
```

```python
# Count EV Registrations By County
pipeline = [
    {"$group": {"_id": "$county", "total": {"$sum": 1}}},
    {"$sort": {"total": -1}}
]
results = db["vehicle_data"].aggregate(pipeline)
for result in results:
    print(result)
```

```python
pipeline = [
    {"$group": {"_id": "$ev_type", "avg_range": {"$avg": "$electric_range"}}}
]
results = db["vehicle_data"].aggregate(pipeline)
for result in results:
    print(result)
```

```python
#Most Popular Models and Makers

pipeline = [
    {
        "$group": {
            "_id": {"make": "$make", "model": "$model"},  # Group by both make and model
            "count": {"$sum": 1}  # Count the number of registrations
        }
    },
    {"$sort": {"count": -1}}  # Sort in descending order of count
]

results = db["vehicle_data"].aggregate(pipeline)

# Print results
print("Popular EV Models:")
for result in results:
    print(result)
```

```python
# Aggregate data
pipeline = [
    {"$group": {"_id": "$cafv_type", "total": {"$sum": 1}}}
]

results = db["vehicle_data"].aggregate(pipeline)

# Print results
print("CAFV Eligibility Analysis:")
for result in results:
    print(result)
```

```python
pipeline = [
    {"$group": {"_id": {"make": "$make", "ev_type": "$ev_type"}, "count": {"$sum": 1}}},
    {"$sort": {"count": -1}},
    {"$group": {
        "_id": "$_id.make",
        "top_ev_type": {"$first": "$_id.ev_type"},
        "count": {"$first": "$count"}
    }}
]

results = db['vehicle_data'].aggregate(pipeline)
for result in results:
    print(result)
```
